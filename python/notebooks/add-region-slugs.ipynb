```python
import ndjson
from slugify import slugify

with open("input.ndjson") as f:
    data = ndjson.load(f)

for row in data:
    if row["_type"] != "region":
        continue
    row["slug"] = {
        "_type": "slug",
        "current": slugify(row["name"]),
    }

with open("output.ndjson", "w") as f:
    ndjson.dump(data, f)
```
